# Estudo de Caso: Ressonância em Circuitos RLC Série
Este notebook faz parte do meu **Guia Pessoal de Engenharia**. O objetivo é calcular a frequência de ressonância e plotar a curva de impedância.

### Fundamentação Teórica
A frequência de ressonância ($f_r$) ocorre quando a reatância indutiva se iguala à capacitiva ($X_L = X_C$):

$$f_r = \frac{1}{2\pi\sqrt{LC}}$$

Neste ponto, a impedância do circuito é puramente **resistiva**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Configurações de visualização para engenharia
plt.style.use('bmh')
```

```python
# 1. Parâmetros do Componente
R = 10       # Ohms
L = 0.1      # Henry
C = 100e-6   # 100 uF (Farads)

# 2. Cálculo da Frequência de Ressonância
fr = 1 / (2 * np.pi * np.sqrt(L * C))
print(f"Frequência de Ressonância Calculada: {fr:.2f} Hz")

# 3. Gerando vetor de frequências para o gráfico (de 1Hz a 200Hz)
f_vetor = np.linspace(1, 200, 500)
omega = 2 * np.pi * f_vetor

# 4. Cálculo da Impedância Z em função da frequência
xl = omega * L
xc = 1 / (omega * C)
z_modulo = np.sqrt(R**2 + (xl - xc)**2)

# 5. Plotagem do Gráfico
plt.figure(figsize=(10, 5))
plt.plot(f_vetor, z_modulo, label='Impedância |Z|', color='blue')
plt.axvline(fr, color='red', linestyle='--', label=f'Ressonância ({fr:.1f} Hz)')

plt.title('Curva de Impedância do Circuito RLC')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Impedância (Ω)')
plt.legend()
plt.grid(True)
plt.show()
```
