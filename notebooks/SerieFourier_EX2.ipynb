# Resolução Simbólica: Série de Fourier

Este notebook detalha o cálculo da Série de Fourier para a função $f(t)$ definida em partes, conforme o problema apresentado.

## 1. Análise da Função e Parâmetros Iniciais

A função dada é:
$$
f(t) = 
\begin{cases} 
0, & -2 < t < -1 \\
6, & -1 < t < 1 \\
0, & 1 < t < 2 
\end{cases}
$$
Com a condição de repetição $f(t+4) = f(t)$.

A partir dessa definição, extraímos as seguintes informações fundamentais:
* **Período ($T$):** A condição $f(t+4) = f(t)$ indica que a função se repete a cada 4 unidades de tempo. Logo, $T = 4$.
* **Meio-período ($L$):** Utilizado nas fórmulas da série, é a metade do período. Logo, $L = \frac{T}{2} = 2$.
* **Intervalos de Integração:** A fórmula geral pede a integração de $-L$ até $L$ (ou seja, de $-2$ a $2$). Porém, como $f(t) = 0$ nos intervalos $[-2, -1]$ e $[1, 2]$, essas partes da integral resultarão em zero. 
* **Conclusão:** Podemos simplificar todos os nossos cálculos integrando apenas de $-1$ a $1$, onde a função vale a constante $6$.

```python
import sympy as sp
from IPython.display import display, Math, Markdown

# Definindo as variáveis simbólicas que serão usadas nos cálculos
# real=True e integer=True ajudam o SymPy a simplificar melhor as equações
t = sp.Symbol('t', real=True)
n = sp.Symbol('n', integer=True, positive=True)

# Definindo os parâmetros extraídos da análise
T = 4
L = sp.Rational(T, 2)  # Usamos sp.Rational para manter as frações exatas
f_val = 6
```

## 2. Cálculo do Coeficiente $a_0$

O coeficiente $a_0$ está relacionado ao nível contínuo (nível DC) ou valor médio da função ao longo de um período. A fórmula matemática é:

$$a_0 = \frac{1}{L} \int_{-L}^{L} f(t) \, dt$$

Aplicando os limites simplificados que deduzimos na análise inicial:
$$a_0 = \frac{1}{2} \int_{-1}^{1} 6 \, dt$$

```python
# Montando a integral de -1 a 1 para o valor f(t) = 6
a0_int = sp.Integral(f_val, (t, -1, 1))
a0_expr = (1/L) * a0_int

# .doit() executa a integral que acabou de ser montada
a0_res = a0_expr.doit()

display(Math(r"a_0 = " + sp.latex(a0_res)))
```

## 3. Cálculo dos Coeficientes $a_n$

Os coeficientes $a_n$ multiplicam os termos de cosseno na série, representando as frequências harmônicas pares. A fórmula geral é:

$$a_n = \frac{1}{L} \int_{-L}^{L} f(t) \cos\left(\frac{n\pi t}{L}\right) dt$$

Substituindo $L=2$ e restringindo a integral ao trecho não nulo da função:
$$a_n = \frac{1}{2} \int_{-1}^{1} 6 \cos\left(\frac{n\pi t}{2}\right) dt$$

```python
# Montando a integral para an
an_int = sp.Integral(f_val * sp.cos(n * sp.pi * t / L), (t, -1, 1))
an_expr = (1/L) * an_int

# Executando e simplificando o resultado trigonométrico
an_res = sp.simplify(an_expr.doit())

display(Math(r"a_n = " + sp.latex(an_res)))
```

## 4. Cálculo dos Coeficientes $b_n$

Os coeficientes $b_n$ multiplicam os termos de seno. A fórmula geral é:

$$b_n = \frac{1}{L} \int_{-L}^{L} f(t) \sin\left(\frac{n\pi t}{L}\right) dt$$

**Dica de Ouro (Paridade da Função):** Observe que $f(t)$ é perfeitamente simétrica em relação ao eixo y (o trecho da esquerda é um espelho do trecho da direita). Isso classifica $f(t)$ como uma **função par**. 
A função seno, por outro lado, é uma **função ímpar**. O produto de uma função par por uma função ímpar resulta em uma função ímpar. 
Uma das propriedades fundamentais das integrais diz que a integral de uma função ímpar sobre um intervalo simétrico (como $[-1, 1]$) é sempre **zero**. 

Portanto, sabemos de antemão que $b_n = 0$. Vamos usar o código abaixo apenas para comprovar isso matematicamente.

```python
# Montando a integral para bn
bn_int = sp.Integral(f_val * sp.sin(n * sp.pi * t / L), (t, -1, 1))
bn_expr = (1/L) * bn_int

# Executando a integral. O SymPy deve retornar 0.
bn_res = sp.simplify(bn_expr.doit())

display(Math(r"b_n = " + sp.latex(bn_res)))
```

## 5. Montagem da Série de Fourier Final

A equação geral da Série de Fourier é dada por:

$$f(t) = \frac{a_0}{2} + \sum_{n=1}^{\infty} \left[ a_n \cos\left(\frac{n\pi t}{L}\right) + b_n \sin\left(\frac{n\pi t}{L}\right) \right]$$

Como provamos que $b_n = 0$, a parte do seno desaparece (Série de Cossenos). Basta substituir os valores de $a_0$ e $a_n$ que encontramos para obter a resposta final do exercício.

```python
# Termo constante da série (a0 / 2)
a0_term = a0_res / 2

# Exibindo a equação final formatada
display(Math(r"\text{Série Final:} \quad f(t) = " + sp.latex(a0_term) + r" + \sum_{n=1}^{\infty} \left[ \left(" + sp.latex(an_res) + r"\right) \cos\left(\frac{n\pi t}{2}\right) \right]"))
```

## 6. Expansão até a 5ª Harmônica

Para expandir a série de Fourier, vamos iterar sobre os valores de $n$ de 1 até 5. Para cada valor, substituímos $n$ na expressão do coeficiente $a_n$ que calculamos anteriormente.

Como $b_n = 0$, cada termo da nossa soma será simplesmente:
$$\text{Termo}_n = a_n \cos\left(\frac{n\pi t}{2}\right)$$

A série truncada na quinta harmônica será a soma do valor médio (nível contínuo, $\frac{a_0}{2}$) com os 5 primeiros termos.

```python
# Inicializamos a série expandida com o termo constante (a0 / 2)
serie_expandida = a0_res / 2

display(Markdown("#### Calculando os coeficientes $a_n$:"))

# Loop de n=1 até n=5
for i in range(1, 6):
    # Substitui a variável simbólica 'n' pelo número inteiro 'i'
    an_val = an_res.subs(n, i)
    
    # Exibe o valor do coeficiente atual
    display(Math(rf"a_{i} = " + sp.latex(an_val)))
    
    # Monta o termo da série com o cosseno e soma na equação total
    termo_atual = an_val * sp.cos(i * sp.pi * t / L)
    serie_expandida += termo_atual

display(Markdown("---"))
display(Markdown("#### Série de Fourier expandida até $n=5$:"))

# Exibe a equação da série completamente expandida
display(Math(r"f_5(t) = " + sp.latex(serie_expandida)))
```

## 7. Gráfico Interativo da Série de Fourier

Para visualizar como a soma dos senos e cossenos se aproxima da nossa função original (uma onda quadrada/retangular), vamos plotar o gráfico.

Abaixo, temos um gráfico interativo. Use o controle deslizante **Harmônicas** para alterar a quantidade de termos da série ($n$) e observe o fenômeno de Gibbs nas bordas (aquelas "vibrações" perto de onde a função "pula" de 0 para 6).

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# 1. Função Original (Numérica) para comparação
def f_original(t_array):
    # Faz o tempo t "dar a volta" a cada 4 unidades (Período = 4)
    # Trazendo qualquer valor de t para o intervalo [-2, 2]
    t_mod = np.mod(t_array + 2, 4) - 2
    # Retorna 6 se t estiver entre -1 e 1, caso contrário retorna 0
    return np.where((t_mod > -1) & (t_mod < 1), 6, 0)

# 2. Função que plota o gráfico atualizado de acordo com o slider
def plot_fourier(num_harmonics):
    # Começamos a equação apenas com o termo constante (a0 / 2)
    expr = a0_res / 2
    
    # Somamos as harmônicas de 1 até o número escolhido no slider
    for i in range(1, num_harmonics + 1):
        an_val = an_res.subs(n, i)
        expr += an_val * sp.cos(i * sp.pi * t / L)
        
    # sp.lambdify converte a equação simbólica do SymPy em uma função 
    # numérica do NumPy, o que deixa o gráfico muito mais rápido.
    f_aprox = sp.lambdify(t, expr, 'numpy')
    
    # Criamos 1000 pontos de tempo (t) entre -4 e 4 (2 períodos completos)
    t_vals = np.linspace(-4, 4, 1000)
    y_orig = f_original(t_vals)
    
    # Calculamos os valores aproximados. 
    # Se num_harmonics=0, f_aprox retorna um número fixo, então forçamos a virar um array
    y_aprox = f_aprox(t_vals)
    if np.isscalar(y_aprox):
        y_aprox = y_aprox * np.ones_like(t_vals)

    # 3. Desenhando o gráfico
    plt.figure(figsize=(10, 5))
    
    # Plota a função original em preto tracejado
    plt.plot(t_vals, y_orig, 'k--', label='Função Original f(t)', linewidth=2)
    
    # Plota a aproximação de Fourier em vermelho
    plt.plot(t_vals, y_aprox, 'r-', label=f'Série de Fourier (n={num_harmonics})', linewidth=1.5)
    
    plt.title(f'Aproximação da Série de Fourier com {num_harmonics} harmônicas')
    plt.xlabel('Tempo (t)')
    plt.ylabel('Amplitude f(t)')
    plt.ylim(-2, 8) # Trava o eixo Y para o gráfico não pular quando o slider mexer
    plt.xlim(-4, 4)
    plt.grid(True)
    plt.legend(loc='upper right')
    
    plt.show()

# 4. Cria o slider interativo e conecta com a função de plotagem
widgets.interact(plot_fourier, 
                 num_harmonics=widgets.IntSlider(min=0, max=50, step=1, value=3, 
                                                 description='Harmônicas:'))
```
