### Estudando a biblioteca Numpy
* Inicializando bibliotecas e parâmetros

```python
import numpy as np
# Define 3 casas decimais e impede notação científica para números pequenos
np.set_printoptions(precision=3, suppress=True)
```

### 1. Criação de Arrays
O `ndarray` é o objeto central do NumPy. Ele permite armazenar dados de forma homogênea e eficiente.
* `np.array()`: Converte listas em arrays.
* `np.zeros()` / `np.ones()`: Cria matrizes preenchidas com 0 ou 1.
* `np.linspace()`: Cria sequências com intervalos calculados.

```python
# Criando a partir de listas
arr_1d = np.array([1, 2, 3])
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Arrays estruturados
zeros = np.zeros((2, 3)) 
espacado = np.linspace(0, 10, 5) # 5 números entre 0 e 10

print("Array 2D:\n", arr_2d)
print("Zeros:\n", zeros)
print("Linspace:", espacado)
```

### 2. Matemática Vetorizada
Diferente das listas, o NumPy permite operações aritméticas diretas sem a necessidade de loops `for`. Isso é conhecido como **vetorização**.

```python
a = np.array([10, 20, 30, 40])
b = np.arange(4) # [0, 1, 2, 3]

print("Soma:", a + b)
print("Multiplicação por escalar:", a * 2)
print("Exponencial:", np.exp(b))
print("Seno:", np.sin(a))
```

### 3. Indexação e Filtros
Acessar dados no NumPy é potente: podemos usar intervalos, índices negativos ou até **máscaras booleanas** para filtrar valores.

```python
matriz = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])

# Slicing: [linhas, colunas]
# Pega as duas primeiras linhas e a última coluna
print("Matriz:\n",matriz)
print("Corte específico:\n", matriz[:2, -1])

# Filtro Booleano
mask = matriz > 50
print("Valores maiores que 50:", matriz[mask])
```

### 4. Manipulação de Formatos e Eixos
Muitas funções aceitam o argumento `axis`:
* `axis=0`: Operação ao longo das colunas (vertical).
* `axis=1`: Operação ao longo das linhas (horizontal).

```python
arr = np.arange(12)
# Reorganiza para 3 linhas e 4 colunas
matriz_3x4 = arr.reshape(3, 4)

print("Matriz:\n", matriz_3x4)
print("Soma de cada coluna (Axis 0):", matriz_3x4.sum(axis=0))
print("Média de cada linha (Axis 1):", matriz_3x4.mean(axis=1))
```

### 5. Estatística e Aleatoriedade
O submódulo `random` é essencial para simulações e inicialização de pesos em modelos de IA.

```python
# Gerando 5 números de uma distribuição normal
normal = np.random.randn(5)

# Métricas rápidas
print("Dados aleatórios:", (normal))
print("Ordem crescente:", np.sort(normal))
print(f"Desvio Padrão:, {np.std(normal):.3f}")
print(f"Variância:, {np.var(normal):.3f}")
print(f"Índice do valor máximo:, {np.argmax(normal):.3f}")
print(f"Valor máximo:, {np.max(normal):.3f}")
print(f"Valor mínimo:, {np.min(normal):.3f}")
print(f"Valor médio:, {np.mean(normal):.3f}")
print(f"Valor da mediana:, {np.median(normal):.3f}")
```

### 6. Broadcasting
O **Broadcasting** é uma funcionalidade poderosa que permite ao NumPy realizar operações aritméticas entre arrays de diferentes formatos (shapes), desde que sejam compatíveis. O array menor é "transmitido" sobre o maior.

```python
import numpy as np

# Um array 2D e um array 1D
matriz = np.array([[1, 2, 3], [4, 5, 6]])
vetor = np.array([10, 20, 30])

# O vetor é somado a cada linha da matriz automaticamente
resultado = matriz + vetor

print("Matriz (2x3):\n", matriz)
print("Vetor (1x3):", vetor)
print("Resultado do Broadcasting:\n", resultado)
```

### 7. Álgebra Linear
O NumPy é amplamente utilizado para cálculos matemáticos complexos. O submódulo `linalg` oferece ferramentas para resolver sistemas, calcular determinantes e muito mais.
* `np.dot()` ou `@`: Multiplicação de matrizes.
* `np.linalg.inv()`: Inversa de uma matriz.
* `np.linalg.det()`: Determinante.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Multiplicação de matrizes (Diferente de A * B)
produto_matriz = np.dot(A, B) # Ou: A @ B

# Inversa e Determinante
inversa_A = np.linalg.inv(A)
determinante_A = np.linalg.det(A)

print("Matriz A:\n", A)
print("Matriz B:\n", B)
print("Produto Matricial:\n", produto_matriz)
print("Inversa:\n", inversa_A)
print(f"Determinante de A:, {determinante_A:.3f}")
```

### 8. Operações com Strings
O NumPy também possui um módulo específico (`char`) para manipular arrays de texto de forma vetorizada.

```python
nomes = np.array(['python', 'numpy', 'jupyter'], dtype=str)

# Transformar em maiúsculas
maiusculo = np.char.upper(nomes)

# Adicionar sufixo
com_sufixo = np.char.add(nomes, '_dev')

# Adicionar prefixo
com_prefixo = np.char.add('pre_',nomes)

print("Original:", nomes)
print("Maiúsculo:", maiusculo)
print("Com Sufixo:", com_sufixo)
print("Com Prefixo:", com_prefixo)
```

### 9. Tratamento de Dados Ausentes
Na ciência de dados, é comum encontrar valores nulos (`NaN` - Not a Number). O NumPy oferece funções específicas que ignoram esses valores em cálculos.

```python
dados = np.array([1, 2, np.nan, 4, 5])

print("vetor:", dados)
print("Soma comum (retorna NaN):", np.sum(dados))
print("Soma ignorando NaN:", np.nansum(dados))
print("Média ignorando NaN:", np.nanmean(dados))
```

### 10. Persistência de Dados
Você pode salvar seus arrays em arquivos binários compactos (`.npy`) ou arquivos de texto (`.txt`, `.csv`).

```python
meus_dados = np.random.rand(5, 5)

# Salvar em formato binário NumPy
np.save('dados_projeto.npy', meus_dados)

# Carregar de volta
dados_carregados = np.load('dados_projeto.npy')

# Salvar como CSV simples
np.savetxt('dados.csv', meus_dados, delimiter=';', fmt='%.3f')

print("Dados salvos e carregados com sucesso!")
```

### 11. Customização da Exibição
Por padrão, o NumPy pode exibir números muito pequenos em notação científica (ex: `7.64e-01`). Podemos configurar o notebook para mostrar apenas um número fixo de casas decimais e suprimir a notação científica para melhorar a legibilidade.

```python
import numpy as np

# Criando dados similares aos da imagem (entre 0 e 1)
dados_exemplo = np.random.rand(5, 5)

# Configurando o NumPy:
# precision=3: Define 3 casas decimais
# suppress=True: Impede o uso de notação científica (e-01, etc)
np.set_printoptions(precision=3, suppress=True)

print("Dados formatados com 3 casas decimais:")
print(dados_exemplo)

# Para resetar as configurações para o padrão original, caso precise:
# np.set_printoptions(edgeitems=3, infstr='inf', linewidth=75, nanstr='nan', 
#                     precision=8, suppress=False, threshold=1000, formatter=None)
```

### 12. Arredondamento Real
Diferente da configuração de exibição acima (que apenas muda como você **vê** o dado), as funções abaixo alteram os **valores reais** dentro do array.

```python
# Criando um array com muitas casas decimais
arr_preciso = np.array([0.76490596, 0.85609321, 0.36945348])

# Arredondando para 3 casas
arr_arredondado = np.around(arr_preciso, decimals=3)

print("Original:", arr_preciso)
print("Arredondado de fato:", arr_arredondado)
```

### 13. Exportação com Precisão Definida
Ao salvar dados para ferramentas externas (como Excel), é comum querer controlar a precisão. 
O argumento `fmt='%.3f'` no `np.savetxt` garante que os dados sejam gravados com apenas 3 casas decimais e sem notação científica.

```python
import numpy as np

# Gerando os dados
meus_dados = np.random.rand(5, 5)

# 1. Configura a exibição no console do Jupyter (não afeta o arquivo salvo)
np.set_printoptions(precision=3, suppress=True)

# 2. Salva em formato binário (mantém a precisão total para uso no Python)
np.save('dados_projeto.npy', meus_dados)

# 3. Salva como CSV formatado (perfeito para leitura humana e Excel)
# %.3f significa: número de ponto flutuante com 3 casas decimais
np.savetxt('dados.csv', meus_dados, delimiter=';', fmt='%.3f')

print("Dados exibidos no console com set_printoptions:")
print(meus_dados)
print("\nArquivo 'dados.csv' gerado com sucesso com 3 casas decimais!")
```
