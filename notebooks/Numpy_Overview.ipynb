### 1. Criação de Arrays
O `ndarray` é o objeto central do NumPy. Ele permite armazenar dados de forma homogênea e eficiente.
* `np.array()`: Converte listas em arrays.
* `np.zeros()` / `np.ones()`: Cria matrizes preenchidas com 0 ou 1.
* `np.linspace()`: Cria sequências com intervalos calculados.

```python
import numpy as np

# Criando a partir de listas
arr_1d = np.array([1, 2, 3])
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Arrays estruturados
zeros = np.zeros((2, 3)) 
espacado = np.linspace(0, 10, 5) # 5 números entre 0 e 10

print("Array 2D:\n", arr_2d)
print("Zeros:\n", zeros)
print("Linspace:", espacado)
```

### 2. Matemática Vetorizada
Diferente das listas, o NumPy permite operações aritméticas diretas sem a necessidade de loops `for`. Isso é conhecido como **vetorização**.

```python
a = np.array([10, 20, 30, 40])
b = np.arange(4) # [0, 1, 2, 3]

print("Soma:", a + b)
print("Multiplicação por escalar:", a * 2)
print("Exponencial:", np.exp(b))
print("Seno:", np.sin(a))
```
